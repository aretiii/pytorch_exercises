```python
import torch

# ===========================
# Exercise 1: Tensor Creation and Operations
# ===========================
shape = (3,2)
rand_tensor1 = torch.rand(shape)
rand_tensor2 = torch.rand(shape)
shape1 = (2,3)
rand_tensor3 = torch.rand(shape1)

print(rand_tensor1)
print(rand_tensor2)
print(rand_tensor3)

# Matrix multiplication
mult_tensor = torch.matmul(rand_tensor1, rand_tensor3)
print(mult_tensor)

# ===========================
# Exercise 2: Element-wise Operations
# ===========================
shape = (4,4)
rand_tensor4 = torch.rand(shape)
print(rand_tensor4)
result = rand_tensor4 - 5
print(result)
squared_result = result ** 2
print(squared_result)

# ===========================
# Exercise 3: Reshaping
# ===========================
tensor = torch.rand(2, 3, 4)
reshaped_tensor = tensor.reshape(6, 4)
print(reshaped_tensor)

# ===========================
# Exercise 4: Broadcasting
# ===========================
tensor1 = torch.rand(3,1)
tensor2 = torch.rand(1,3)
result = tensor1 + tensor2
print(result)

# ===========================
# Exercise 5: Concatenation
# ===========================
tensor3 = torch.rand(3,)
tensor4 = torch.rand(3,)
tensor5 = torch.rand(3,)
t1 = torch.cat([tensor3, tensor4, tensor5], dim=0)
print(t1)

# ===========================
# Exercise 6: Indexing and Masking
# ===========================
tensor6 = torch.rand(4,4)
print(tensor6)
tensor6[:,1] = 0
print(tensor6)
mask = tensor > 0.5
extracted_values = tensor[mask]
print(extracted_values)

# ===========================
# Exercise 7: Gradients
# ===========================
x = torch.randn(3, 3, requires_grad=True)
print("Original tensor:\n", x)

y = x * 2 + 3
print("\nResult after operation:\n", y)

y.backward(torch.ones_like(y))
print("\nGradient of y with respect to x:\n", x.grad)
```
